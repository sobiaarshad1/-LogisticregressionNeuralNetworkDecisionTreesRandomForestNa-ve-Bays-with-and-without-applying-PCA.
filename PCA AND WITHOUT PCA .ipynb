```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import warnings
warnings.filterwarnings("ignore")
import os
```

```python
#load the dataset
df = pd.read_csv('breast-cancer.csv')
df
# replacing values
df['diagnosis'].replace(['M', 'B'],
                        [1, 0], inplace=True)
```

```python
breast=df.drop(["id"],axis=1)
breast
```

```python
#x for data without target
x=breast.drop(["diagnosis"],axis=1)
x=x.fillna(0) 
x
```

```python
#here select the target without data
target=breast["diagnosis"]
target
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
# Scale the features (if required)
scaler = StandardScaler()
X = scaler.fit_transform(x)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x, target, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

# Create and train the logistic regression model
logreg = LogisticRegression()
logreg.fit(X_train, y_train)

# Predict on the test set
y_pred_logreg = logreg.predict(X_test)
print(y_pred_logreg)
```

```python
from sklearn.neural_network import MLPClassifier

# Create and train the neural network model
nn = MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=1000, random_state=42)
nn.fit(X_train, y_train)

# Predict on the test set
y_pred_nn = nn.predict(X_test)
print(y_pred_nn)
```

```python
from sklearn.tree import DecisionTreeClassifier

# Create and train the decision tree model
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)

# Predict on the test set
y_pred_dt = dt.predict(X_test)
print(y_pred_dt)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Create and train the random forest model
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Predict on the test set
y_pred_rf = rf.predict(X_test)

print(y_pred_rf)
```

```python
from sklearn.naive_bayes import GaussianNB

# Create and train the Naïve Bayes model
nb = GaussianNB()
nb.fit(X_train, y_train)

# Predict on the test set
y_pred_nb = nb.predict(X_test)
print(y_pred_nb)
```

PCA IMPLEMENTATION

```python
from sklearn.decomposition import PCA

# Apply PCA on the training data
pca = PCA(n_components=10)  # Choose the desired number of components
X_train_pca = pca.fit_transform(X_train)

# Transform the test data using the fitted PCA
X_test_pca = pca.transform(X_test)
```

LOGISTIC REGRESSION WITH PCA 

```python
from sklearn.linear_model import LogisticRegression

# Create and train the logistic regression model
logreg = LogisticRegression()
logreg.fit(X_train_pca, y_train)

# Predict on the test set
y_pred_logreg = logreg.predict(X_test_pca)
print(y_pred_logreg)
```

NEURAL NETWORK WITH PCA 

```python
from sklearn.neural_network import MLPClassifier

# Create and train the neural network model
nn = MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=1000, random_state=42)
nn.fit(X_train_pca, y_train)

# Predict on the test set
y_pred_nn = nn.predict(X_test_pca)
print(y_pred_nn)
```

DecisionTreeClassifier WITH PCA

```python
from sklearn.tree import DecisionTreeClassifier

# Create and train the decision tree model
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train_pca, y_train)

# Predict on the test set
y_pred_dt = dt.predict(X_test_pca)
print(y_pred_dt)
```

RANDOM FOREST CLASSIFER WITH PCA 

```python
from sklearn.ensemble import RandomForestClassifier

# Create and train the random forest model
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train_pca, y_train)

# Predict on the test set
y_pred_rf = rf.predict(X_test_pca)

print(y_pred_rf)
```

NAIVE BAYES WITH PCA

```python
from sklearn.naive_bayes import GaussianNB

# Create and train the Naïve Bayes model
nb = GaussianNB()
nb.fit(X_train_pca, y_train)

# Predict on the test set
y_pred_nb = nb.predict(X_test_pca)
print(y_pred_nb)
```

PRECISION WITHOUT PCA

```python
from sklearn.metrics import precision_score, recall_score

# Calculate precision and recall for Logistic Regression without PCA
precision_logreg = precision_score(y_test, y_pred_logreg)
print(precision_logreg)
recall_logreg = recall_score(y_test, y_pred_logreg)
print(recall_logreg)
```

```python
# Calculate precision and recall for Neural Network without PCA
precision_nn = precision_score(y_test, y_pred_nn)
print(precision_nn)
recall_nn = recall_score(y_test, y_pred_nn)
print(recall_nn)
```

```python
# Calculate precision and recall for Decision Trees without PCA
precision_dt = precision_score(y_test, y_pred_dt)
print(precision_dt )
recall_dt = recall_score(y_test, y_pred_dt)
print(recall_dt)
```

```python
# Calculate precision and recall for Random Forest without PCA
precision_rf = precision_score(y_test, y_pred_rf)
print(precision_rf)
recall_rf = recall_score(y_test, y_pred_rf)
print(recall_rf)
```

```python
# Calculate precision and recall for Naïve Bayes without PCA
precision_nb = precision_score(y_test, y_pred_nb)
print(precision_nb)
recall_nb = recall_score(y_test, y_pred_nb)
print(recall_nb)
```

PRECISION WITH PCA

```python
# Calculate precision and recall for Logistic Regression with PCA
y_pred_logreg_pca = logreg.predict(X_test_pca)
precision_logreg_pca = precision_score(y_test, y_pred_logreg_pca)
print(precision_logreg_pca)
recall_logreg_pca = recall_score(y_test, y_pred_logreg_pca)
print(recall_logreg_pca )
```

```python
# Calculate precision and recall for Neural Network with PCA
y_pred_nn_pca = nn.predict(X_test_pca)
precision_nn_pca = precision_score(y_test, y_pred_nn_pca)
print(precision_nn_pca )
recall_nn_pca = recall_score(y_test, y_pred_nn_pca)
print(recall_nn_pca)
```

```python
# Calculate precision and recall for Decision Trees with PCA
y_pred_dt_pca = dt.predict(X_test_pca)
precision_dt_pca = precision_score(y_test, y_pred_dt_pca)
print(precision_dt_pca)
recall_dt_pca = recall_score(y_test, y_pred_dt_pca)
print(recall_dt_pca )
```

```python

# Calculate precision and recall for Random Forest with PCA
y_pred_rf_pca = rf.predict(X_test_pca)
precision_rf_pca = precision_score(y_test, y_pred_rf_pca)
print(precision_rf_pca)
recall_rf_pca = recall_score(y_test, y_pred_rf_pca)
print(recall_rf_pca)
```

```python
# Calculate precision and recall for Naïve Bayes with PCA
y_pred_nb_pca = nb.predict(X_test_pca)
precision_nb_pca = precision_score(y_test, y_pred_nb_pca)
print(precision_nb_pca)
recall_nb_pca = recall_score(y_test, y_pred_nb_pca)
print(recall_nb_pca)
```
